###Lambda expressions
###Map
###Filter

```
#map function
def square(num):
  return num**2
```

```
my_nums=[1,2,3,4,5]
for item in map(square,my_nums):
  print(item)
```

```
list(map(square,my_nums))
```

```
def splicer(mystring):
  if len(mystring)%2 ==0:
    return "EVEN"
  else:
    return mystring[0]
```

```
names = ['Andy','Eve','Sally']
```

```
list(map(splicer,names))
```

```
#Filter function: picks value on boolean value
def check_even(num):
  return num%2 ==0
```

```
mynums = [1,2,3,4,5,6]
```

```
list(filter(check_even,mynums))
```

```
#lambda func: Those functions that will be used only once, function call is not be feasible. Lambda function are called there.
def square(num):
  return num**2
```

```
square(3)
```

```
square = lambda num: num**2
```

```
square(3)
```

```
list(map(lambda num: num**2,mynums))
```

```
list(filter(lambda num:num%2 ==0,mynums))
```

```
names
```

```
list(map(lambda name:name[0],names))
```

```
#reverse
list(map(lambda x:x[::-1],names))
```
