###List comprehension: It is a unique way to creat a list. To creat a for loop along with .append() to creat a list, then it is used.

```
str1 = 'hello'
```

```
#'append' adds element in the last of a list
list1 = [1,2]
for letter in str1:
  # list1 = []
  list1.append(letter)
print(list1)
```

```
#list comprhension
list1 = [letter for letter in str1]
list1
```

```
list2 = [x for x in 'swati']
list2
```

```
list3 = [num for num in range(0,10)]
list3
```

```
list3 = [num**2 for num in range(0,10)]
list3
```

```
list4 = [num for num in range(0,10) if num%2==0]
list4
```

```
#conversion using 'list comprehension'
celcious = [0,10,20,34.5]

farenheit = [((9/5)*temp + 32) for temp in celcious]
farenheit
```

```
#conversion using for loop
celcious = [0,10,20,34.5]

farenheit = []
for temp in celcious:
  farenheit.append((9/5)*temp + 32)
farenheit
```

```
mylist = []

for x in [2,4,6]:
  for y in [100,200,300]:
    # print(y)
    # print(x)
    mylist.append(x*y)

mylist
```
